```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
Observation:
More passengers died than survived in the Titanic disaster.
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
Observation:
Females had significantly higher survival rate than males.
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
Observation:
Passengers in 1st class survived more compared to 2nd and 3rd class.
```

```python
sns.histplot(df['Age'], bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
Observation:
Most passengers were aged between 2040 years.
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Observation:
Passenger class and fare show strong relation with survival.
```

```python
Final Conclusion:
Gender influenced survival
Higher class passengers survived more
Age and fare impacted survival chances
```
